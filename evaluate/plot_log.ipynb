```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Times New Roman'
```

```python
logs = pd.read_csv(r"path/to/results/train.csv")
```

```python
loss = logs.loss
val_loss = logs.val_loss
epochs = list(range(1, len(loss)+1))
```

```python
fig, ax = plt.subplots(figsize=(10,10), dpi=200)
ax.plot(epochs, loss, 'g', label='loss')
ax.plot(epochs, val_loss, 'r', label='val_loss')
ax.set_ylim([-1, None])
ax.legend(fontsize=20)
ax.set_xlabel('Epochs', fontsize=15)
ax.set_ylabel('Cosine Similarity Loss', fontsize=15)
plt.savefig('converge.png', facecolor='white', transparent=False)
```
